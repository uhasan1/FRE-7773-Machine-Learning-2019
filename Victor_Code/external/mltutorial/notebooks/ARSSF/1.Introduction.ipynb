**These are the basic rules:**
- You score a point once your saber touches the other player upper body.
- If both players hit with in 120ms from each other, the player with the Right of Way (RoW) gets the point. In simpler terms the attacker gets the point.
- If both attacked at the same time and and hit each other within 120ms no point is awarded to neither one of them.
- Players loose the RoW once they are blocked or if they miss.
- There are other more rules...

## Inspecting the properties of `VideoCapture` Object

propId Reference:

http://docs.opencv.org/modules/highgui/doc/reading_and_writing_images_and_video.html#videocapture-get

`VideoCapture.get(propId) -> retval`

```
propId
    CV_CAP_PROP_POS_MSEC Current position of the video file in milliseconds or video capture timestamp.
    CV_CAP_PROP_POS_FRAMES 0-based index of the frame to be decoded/captured next.
    CV_CAP_PROP_POS_AVI_RATIO Relative position of the video file: 0 - start of the film, 1 - end of the film.
    CV_CAP_PROP_FRAME_WIDTH Width of the frames in the video stream.
    CV_CAP_PROP_FRAME_HEIGHT Height of the frames in the video stream.
    CV_CAP_PROP_FPS Frame rate.
    CV_CAP_PROP_FOURCC 4-character code of codec.
    CV_CAP_PROP_FRAME_COUNT Number of frames in the video file.
    CV_CAP_PROP_FORMAT Format of the Mat objects returned by retrieve() .
    CV_CAP_PROP_MODE Backend-specific value indicating the current capture mode.
    CV_CAP_PROP_BRIGHTNESS Brightness of the image (only for cameras).
    CV_CAP_PROP_CONTRAST Contrast of the image (only for cameras).
    CV_CAP_PROP_SATURATION Saturation of the image (only for cameras).
    CV_CAP_PROP_HUE Hue of the image (only for cameras).

```

````python
propId_dict = {"POS_MSEC": 0,
"POS_FRAMES": 1,
"POS_AVI_RATIO": 2,
"FRAME_WIDTH": 3,
"FRAME_HEIGHT": 4,
"FPS": 5,
"FOURCC": 6,
"FRAME_COUNT": 7,
"FORMAT": 8,
"MODE": 9,
"BRIGHTNESS": 10,
"CONTRAST": 11,
"SATURATION": 12,
"HUE": 13,
"GAIN": 14,
"EXPOSURE": 15,
"CONVERT_RGB": 16,
"WHITE_BALANCE": 17,
"RECTIFICATION": 18,
}
````

````python
print cap.get(1)
````

````python
for key,value in sorted(propId_dict.iteritems(), key=lambda item: item[1], reverse=False):
    print "%s. %s: %s" % (value, key, cap.get(value))
````

### FourCC

FourCC is a 4 digit adentifier for a codec.

````python
def get_fourcc(cap):
    int_value = cap.get(propId_dict["FOURCC"])
    format_hex = format(int(int_value), "x")
    fourcc = ""
    for index in range(len(format_hex))[::2]:
        value = format_hex[index:index+2]
        fourcc = chr(int(value, 16)) + fourcc
    return fourcc
get_fourcc(cap)
````

### Video Length & Current Position

````python
def get_video_length(cap, current_pos=False, in_ms=False):
    """
    Measure the length of a VideoCapture Object
    
    Parameters:
    ===========
    cap: VideoCapture object
    in_ms: boolean (default=False)
      Returns the time in milliseconds if set to True.
    current_pos: boolean (default=False)
      Returns the time for the current position
      
    Return:
    =======
    if in_ms=False -> str
      example for 1 second VideoCapture:
      > get_video_length(cap, in_ms=False)
      00:00:01.0
    
    if in_ms=True  -> int
      example for 1 second VideoCapture
      > get_video_length(cap, in_ms=True)
      1000
    """
    fps = cap.get(propId_dict["FPS"])
    if current_pos:
        length_in_ms = cap.get(propId_dict["POS_MSEC"])
    else:
        frame_count = cap.get(propId_dict["FRAME_COUNT"])
        length_in_ms = int(frame_count/fps * 1000)
    if in_ms:
        return length_in_ms
    length = datetime.timedelta(milliseconds=length_in_ms)
    return length

print get_video_length(cap,current_pos=True)
````

## Matplotlib

````python
def show_image(frame=0):
    # Set Frame
    cap.set(propId_dict["POS_FRAMES"], frame)
    
    dpi=72.0
    height = cap.get(propId_dict["FRAME_HEIGHT"])
    width = cap.get(propId_dict["FRAME_WIDTH"])

    plt.figure(figsize=(width/dpi,height/dpi))
    ret, image_bgr = cap.read()
    image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
    plt.imshow(image)

    plt.grid()
    plt.show()
show_image(frame=130)
````

````python
total_frames = cap.get(propId_dict["FRAME_COUNT"])
print total_frames
interact(show_image, frame=[130,500])
````
