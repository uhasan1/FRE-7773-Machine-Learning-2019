# Building Work Interface

```python
roi_list = []
```

```python
def on_mouse(event, x, y, botton, *args):
    if event == 1 & botton == 1: #mouse down
        roi_list.append([x,y,botton])
    return None

def show_frame(param):
    frame = cv2.getTrackbarPos("frame", "Saber Fencing")
    cap.set(propId_dict["POS_FRAMES"], frame)
    ret, image_bgr = cap.read()
    
    color_schema = cv2.getTrackbarPos("BGR/HSV", "Saber Fencing")
    color_schema = {0:"bgr", 1:"hsv"}[color_schema]
    
    imge_or_threshold = cv2.getTrackbarPos("Image/Threshold", "Saber Fencing")
    imge_or_threshold = {0:"image",1:"threshold"}[imge_or_threshold]
    
    
    max_b = cv2.getTrackbarPos("Max B", "BGR Histogram")
    min_b = cv2.getTrackbarPos("Min B", "BGR Histogram")
    max_g = cv2.getTrackbarPos("Max G", "BGR Histogram")
    min_g = cv2.getTrackbarPos("Min G", "BGR Histogram")
    max_r = cv2.getTrackbarPos("Max R", "BGR Histogram")
    min_r = cv2.getTrackbarPos("Min R", "BGR Histogram")
    
    max_h = cv2.getTrackbarPos("Max H", "HSV Histogram")
    min_h = cv2.getTrackbarPos("Min H", "HSV Histogram")
    max_s = cv2.getTrackbarPos("Max S", "HSV Histogram")
    min_s = cv2.getTrackbarPos("Min S", "HSV Histogram")
    max_v = cv2.getTrackbarPos("Max V", "HSV Histogram")
    min_v = cv2.getTrackbarPos("Min V", "HSV Histogram")
    
    if color_schema == "bgr":
        display_image = image_bgr
        if imge_or_threshold == "threshold":
            display_image = cv2.inRange(display_image,
                                        np.array([min_b, min_g, min_r]),
                                        np.array([max_b, max_g, max_r]))
    elif color_schema == "hsv":
        display_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
        if imge_or_threshold == "threshold":
            display_image = cv2.inRange(display_image,
                                        np.array([min_h, min_s, min_v]),
                                        np.array([max_h, max_s, max_v]))
    
    cv2.imshow("Saber Fencing",display_image)
    
    show_hist(image_bgr, limits=np.array([[max_b, min_b], [max_g, min_g], [max_r, min_r]]))
    show_hist(image_bgr, color_schema="hsv", limits=np.array([[max_h, min_h], [max_s, min_s], [max_v, min_v]]))
    return image_bgr

def fig2data (fig):
    fig.canvas.draw ()
 
    w,h = fig.canvas.get_width_height()
    buf = numpy.ndarray(shape = (h, w,4), dtype=numpy.uint8, buffer=fig.canvas.buffer_rgba())
    return buf

def show_hist(image_bgr, color_schema="bgr", limits=np.array([[0,255], [0,255], [0,255]]), color=('b','g','r')):
    fig = plt.figure(figsize=(5,3))
    if color_schema == "bgr":
        image_hist = image_bgr
    elif color_schema == "hsv":
        image_hist = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
        
    for i,col in enumerate(color):
        histr = cv2.calcHist([image_hist],[i],None,[limits[i][0]-limits[i][1]],[limits[i][1],limits[i][0]])
        plt.plot(range(limits[i][1], limits[i][0]), histr, color = col, hold=True)
    plt.xlim([np.min(limits, axis=0)[1], np.max(limits, axis=0)[0]])
    data = fig2data(fig)
    data = cv2.cvtColor(data, cv2.COLOR_RGBA2BGR)
    cv2.imshow(color_schema.upper() + " Histogram",data)
    plt.close()

cv2.namedWindow("Saber Fencing", cv2.WINDOW_NORMAL)
cv2.createTrackbar("frame", "Saber Fencing", 130, int(total_frames), show_frame)
cv2.createTrackbar("BGR/HSV", "Saber Fencing", 0, 1, show_frame)
cv2.createTrackbar("Image/Threshold", "Saber Fencing", 0, 1, show_frame)

cv2.namedWindow("BGR Histogram",  cv2.WINDOW_AUTOSIZE)
cv2.namedWindow("HSV Histogram",  cv2.WINDOW_AUTOSIZE)

cv2.createTrackbar("Max B", "BGR Histogram", 255, 255, show_frame)
cv2.createTrackbar("Min B", "BGR Histogram", 0, 255, show_frame)
cv2.createTrackbar("Max G", "BGR Histogram", 255, 255, show_frame)
cv2.createTrackbar("Min G", "BGR Histogram", 0, 255, show_frame)
cv2.createTrackbar("Max R", "BGR Histogram", 255, 255, show_frame)
cv2.createTrackbar("Min R", "BGR Histogram", 0, 255, show_frame)

cv2.createTrackbar("Max H", "HSV Histogram", 180, 180, show_frame)
cv2.createTrackbar("Min H", "HSV Histogram", 0, 180, show_frame)
cv2.createTrackbar("Max S", "HSV Histogram", 255, 255, show_frame)
cv2.createTrackbar("Min S", "HSV Histogram", 0, 255, show_frame)
cv2.createTrackbar("Max V", "HSV Histogram", 255, 255, show_frame)
cv2.createTrackbar("Min V", "HSV Histogram", 0, 255, show_frame)

image_bgr = show_frame(130)

cv2.moveWindow("Saber Fencing", 10, 25)
cv2.moveWindow("BGR Histogram", 900, 400)
cv2.moveWindow("HSV Histogram", 900, 25)

cv2.setMouseCallback("Saber Fencing", on_mouse)

while True:
    k=cv2.waitKey(0)
    if k==27:
        image_bgr = show_frame(None)
        roi_list = np.array(roi_list)
        image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
        image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
        break
    else:
        print k
    
cv2.destroyAllWindows()
```

# OCR

## Training Set

I have built a training set using Gimp (a photo editing program).

![](image/ocr_digits.png)

I generated the original row using a text tool then replicated it several times. Then I resized each row to a different level and resized them back to their original size.

## Pre-Processing the Image

### Change the image from BGR to gray scale.

```python
training_image = cv2.imread("image/ocr_digits.png")
print "Origianl Shape: %s: " % str(training_image.shape)
training_image = cv2.cvtColor(training_image, cv2.COLOR_BGR2GRAY)
print "Final Shape: %s: " % str(training_image.shape)
plt.imshow(training_image, cmap=plt.cm.Greys)
plt.colorbar()
plt.show()
```

### Threshold the Image

#### Fixed Threshold

```python
fixed_thresh = (training_image < 30) * 255
plt.imshow(fixed_thresh, cmap=plt.cm.Greys)
plt.colorbar()
plt.show()
```

#### Adaptive Threshold

```python
adaptive_thresh = cv2.adaptiveThreshold(training_image,255,1,1,11,2)
plt.imshow(adaptive_thresh, cmap=plt.cm.Greys)
plt.colorbar()
plt.show()
```

## Finding Digits

### Find Contours

```python
contours,hierarchy = cv2.findContours(adaptive_thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

plt.imshow(adaptive_thresh, cmap=plt.cm.Greys)
plt.colorbar()
plt.show()
```

### Filtering out small contours

```python
boxes = []
image_gbr = cv2.cvtColor(training_image, cv2.COLOR_GRAY2RGB)
for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > 100:
        [x,y,w,h] = cv2.boundingRect(cnt)
        if h > 29:
            boxes.append([x,y,w,h])
            cv2.rectangle(image_gbr, pt1=(x,y), pt2=(x+w,y+h), color=[255,0,0], thickness=2)
print len(boxes)
plt.imshow(image_gbr)
```

### Checking boxes - Using Nearest Interpolation

```python
x,y,w,h = boxes[0]
plt.imshow(training_image[y:y+h,x:x+w], cmap=plt.cm.Greys, interpolation="nearest")
plt.colorbar()
plt.show()
```

### Labeling Digits

```python
row = 0
for index in range(row*10,row*10+10):
    x,y,w,h = boxes[index]
    plt.imshow(training_image[y:y+h,x:x+w], cmap=plt.cm.Greys, interpolation="nearest")
    plt.colorbar()
    plt.show()
```

```python
results = []
results.append([9,8,7,6,5,4,3,2,0,1])
results.append([9,8,7,6,5,4,3,2,1,0])
results.append([9,8,6,5,3,2,0,7,4,1])
results.append([9,8,7,6,5,4,3,2,1,0])
results.append([9,8,7,6,5,4,3,2,1,0])
results.append([9,8,7,6,5,4,3,2,1,0])
results = np.ravel(results)
results
```

### Resizing RoI / Normalize

```python
roi = training_image[y:y+h,x:x+w]
roi = roi / 255.
roi = cv2.resize(roi,(10,14))
plt.imshow(roi, cmap=plt.cm.Greys, interpolation="nearest")
plt.colorbar()
plt.show()
```

```python
features = []
for index in range(60):
    x,y,w,h = boxes[index]
    roi = training_image[y:y+h,x:x+w]
    roi = roi / 255.
    roi = cv2.resize(roi,(10,14))
    features.append(np.ravel(roi))
features = np.array(features)
features.shape
```

## Training Classifier

### k-NN

#### Training

```python
knn_cls = KNeighborsClassifier()
param_grid = [{"n_neighbors":range(1,20)}]
knn_gs = GridSearchCV(knn_cls, param_grid=param_grid)
knn_gs.fit(features, results)
knn_cls = knn_gs.best_estimator_
knn_cls
```

#### Testing on training data

```python
knn_cls.fit(features, results)
%timeit knn_cls.predict(features)
knn_cls.score(features, results)
```

## Prediction

```python
def get_boxes(roi):
    boxes = []
    
    contours,hierarchy = cv2.findContours(roi.astype(np.uint8),cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
    
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 100:
            [x,y,w,h] = cv2.boundingRect(cnt)
            if h > 29 and w < 50:
                boxes.append([x,y,w,h])
    return boxes


def read_score(roi, image_hsv_loc=None):
    if not image_hsv_loc is None:
        image_hsv_final = image_hsv_loc
    else:
        image_hsv_final = image_hsv
    roi = get_roi(image_hsv_final, roi)
    roi = ((roi[:,:,2] < 230) * 255).astype(np.uint8)
    
    boxes = get_boxes(roi)
    digits = []
    roi = roi / 255.
    for box in boxes:
        x, y, w, h = box
        digit_roi = cv2.resize(roi[y:y+h,x:x+w],(10,14))
        digits.append(str(knn_cls.predict(np.ravel(digit_roi))[0]))
    if len(boxes) > 1:
        if boxes[0][0] > boxes[1][0]:
            digits = digits[::-1]
    return int("".join(digits))
        
print "Left Score: %s" % read_score(left_player_score_roi)
print "Left Score: %s" % read_score(right_player_score_roi)
```

### Complete Video

```python
def get_image(frame=0):
    # Set Frame
    cap.set(propId_dict["POS_FRAMES"], frame)
    ret, image_bgr = cap.read()
    image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
    return image_hsv

total_frames = cap.get(propId_dict["FRAME_COUNT"])

left_score_list = []
right_score_list = []
for index in range(130, int(total_frames)):
    image_hsv = get_image(index)
    left_score_list.append(read_score(left_player_score_roi, image_hsv))
    right_score_list.append(read_score(right_player_score_roi, image_hsv))
```
